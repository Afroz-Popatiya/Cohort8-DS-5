# Data Cleaning

```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/raw/hour.csv')

print(f'Dataset: {df.shape[0]:,} rows, {df.shape[1]} columns')
print(df.head())
```

# 1. Columns To Drop

```python
cols_to_drop = ['dteday', 'atemp', 'casual', 'registered']

df = df.drop(columns=cols_to_drop)

print(f'Dataset: {df.shape[0]:,} rows, {df.shape[1]} columns')
print(f'Remaining columns: {list(df.columns)}')
```

# 2. Add Variable Labels for Visualization

```python
# Season labels
season_map = {1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'}
df['season_label'] = df['season'].map(season_map)

# Weather labels
weather_map = {1: 'Clear', 2: 'Cloudy', 3: 'Light Rain', 4: 'Heavy Rain'}
df['weather_label'] = df['weathersit'].map(weather_map)

# Year labels
year_map = {0: 2011, 1: 2012}
df['year_label'] = df['yr'].map(year_map)

# Day type labels
day_type_map = {1: 'Working Day', 0: 'Weekend/Holiday'}
df['day_type_label'] = df['workingday'].map(day_type_map)

print(f'Dataset: {df.shape[0]:,} rows, {df.shape[1]} columns')
print('Added column labels:')
print(df[['season', 'season_label', 'weathersit', 'weather_label', 'yr', 'year_label', 'workingday', 'day_type_label']].head())
```

# 3. Decode Weather Values
Temperature, humidity, and wind were scaled 0-1

```python
df['temp_celsius'] = (df['temp'] * 41).round(1)
df['humidity_pct'] = (df['hum'] * 100).round(1)
df['windspeed_kmh'] = (df['windspeed'] * 67).round(1)

print(f'Dataset: {df.shape[0]:,} rows, {df.shape[1]} columns')
print('Added decoded columns for weather variables:')
print(df[['temp', 'temp_celsius', 'hum', 'humidity_pct', 'windspeed', 'windspeed_kmh']].head())
```

# 4. Final Dataset

```python
print(f'Final dataset: {df.shape[0]:,} rows × {df.shape[1]} columns')
print(df.head())
```

# 5. Save Cleaned Dataset

```python
df.to_csv('../data/processed/hour_clean.csv', index=False)
```
